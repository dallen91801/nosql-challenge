### 1. Which establishments have a hygiene score equal to 20?

```python
### 1. Which establishments have a hygiene score equal to 20?
query = {"scores.Hygiene": 20}
# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(count)
# Display the first document in the results using pprint
first_result = establishments.find_one(query)
pprint(first_result)
```

```python
# Convert the result to a Pandas DataFrame
results = list(establishments.find(query))
df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
row_count = df.shape[0]
print(row_count)

# Display the first 10 rows of the DataFrame
df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Define the query to find establishments in London with a RatingValue >= 4
query = {
    "LocalAuthorityName": {"$regex": "London", "$options": "i"},  # Case-insensitive regex for London
    "RatingValue": {"$gte": 4}
}

# Step 1: Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print("Number of establishments with RatingValue >= 4 in London:", count)

# Step 2: Display the first document in the results using pprint
print("First document with RatingValue >= 4 in London:")
pprint(establishments.find_one(query))

# Step 3: Retrieve documents matching the query and convert to a DataFrame
results = list(establishments.find(query))
df_london_ratings = pd.DataFrame(results)

# Step 4: Display the first few rows of the DataFrame
df_london_ratings.head()
# AI was used to verify the results. The first document in the results has a RatingValue of 4 and is located in London.
```

```python
# Retrieve documents matching the query and convert to a DataFrame
results = list(establishments.find(query))
df_london_ratings = pd.DataFrame(results)

# Display the first few rows of the DataFrame
df_london_ratings.head()
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Define the coordinates for "Penang Flavours"
latitude = 51.49014200
longitude = 0.08384000

# Define the degree range for the search
degree_search = 0.05  # Expanding to a larger range for more results

# Define the query with filters for RatingValue, latitude, and longitude range
query = {
    "RatingValue": 5,
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search}
}

# Sort by hygiene score in ascending order and limit to the top 5
sort = [("scores.Hygiene", 1)]
limit = 5

# Execute the query with sorting and limit
results = list(establishments.find(query).sort(sort).limit(limit))

# Print each result
if results:
    print("Top 5 establishments with RatingValue of 5, sorted by lowest hygiene score:")
    for result in results:
        pprint(result)
else:
    print("No establishments found within the specified criteria.")

```

```python
# Convert result to Pandas DataFrame
df_top5 = pd.DataFrame(results)

# Display the DataFrame
df_top5.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Define the aggregation pipeline
pipeline = [
    {"$match": {"scores.Hygiene": 0}},  # Step 1: Match establishments with hygiene score of 0
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},  # Step 2: Group by Local Authority and count
    {"$sort": {"count": -1}}  # Step 3: Sort by count in descending order
]

# Execute the aggregation pipeline
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of Local Authorities with establishments having a hygiene score of 0:", len(results))

# Print the first 10 results
print("Top 10 Local Authorities with the most establishments having a hygiene score of 0:")
pprint(results[:10])
```

```python
# Convert the aggregation result to a DataFrame
df_local_authorities = pd.DataFrame(results)

# Display the number of rows in the DataFrame
num_rows = df_local_authorities.shape[0]
print("Number of rows in the DataFrame:", num_rows)

# Display the first 10 rows of the DataFrame
df_local_authorities.head(10)
```
